##### **Problem 04:**
Generate the vector `[25, 16, 9, 4, 1]` and then find the square root of each element of the vector using for loop. **Hint:** Use `np.sqrt()` function to find the square root of a number, don't use any other built-in function.

```python
# Vector input [25, 16, 9, 4, 1]
vector = np.array([25, 16, 9, 4, 1])

# Square root of each element using a for loop
sqrt_vector = np.zeros_like(vector)  # Create an array of zeros with the same shape as the vector

for i in range(vector.size):
    sqrt_vector[i] = np.sqrt(vector[i])

print("Original Vector:", vector)
print("Square Root Vector:", sqrt_vector)
```

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
# Input x, y values from -3 to 3 with interval 0.5
x = np.arange(-3, 3.5, 0.5)
y = np.arange(-3, 3.5, 0.5)

# Create meshgrid of x, y values
X, Y = np.meshgrid(x, y)

# z values for the equation x + y + z = 1
Z = 1 - X - Y

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Surface Plot of x + y + z = 1')
plt.show()
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivatives
def f(x):
    return np.exp(x)

def f_prime(x):
    return np.exp(x)

def f_double_prime(x):
    return np.exp(x)

# Set the x values from -1 to 1 with an interval of 0.01
x = np.arange(-1, 1.01, 0.01)

# Compute the Taylor approximations
taylor_approx_2x = f(0) + f_prime(0) * x + 0.5 * f_double_prime(0) * x**2
taylor_approx_3x = f(0) + f_prime(0) * x + 0.5 * f_double_prime(0) * x**2
taylor_approx_5x = f(0) + f_prime(0) * x + 0.5 * f_double_prime(0) * x**2

# Compute the original function values
original_func_2x = f(2 * x)
original_func_3x = f(3 * x)
original_func_5x = f(5 * x)

# Calculate the mean squared error (MSE)
mse_2x = np.mean((taylor_approx_2x - original_func_2x)**2)
mse_3x = np.mean((taylor_approx_3x - original_func_3x)**2)
mse_5x = np.mean((taylor_approx_5x - original_func_5x)**2)

# Determine the more accurate method based on MSE
if mse_2x * mse_3x < mse_5x:
    more_accurate_method = "Method 1 (Multiplying Taylor approximations)"
else:
    more_accurate_method = "Method 2 (Taylor approximation for e^5x)"

# Plot the Taylor approximations and the original functions
plt.figure(figsize=(10, 6))
plt.plot(x, taylor_approx_2x, label="Taylor Approximation (e^2x)")
plt.plot(x, taylor_approx_3x, label="Taylor Approximation (e^3x)")
plt.plot(x, taylor_approx_5x, label="Taylor Approximation (e^5x)")
plt.plot(x, original_func_2x, label="Original Function (e^2x)")
plt.plot(x, original_func_3x, label="Original Function (e^3x)")
plt.plot(x, original_func_5x, label="Original Function (e^5x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Taylor Approximation vs Original Function")
plt.legend()
plt.grid(True)
plt.show()

# Print the Mean Squared Error and the more accurate method
print("Mean Squared Error (MSE) - e^2x:", mse_2x)
print("Mean Squared Error (MSE) - e^3x:", mse_3x)
print("Mean Squared Error (MSE) - e^5x:", mse_5x)
print("More Accurate Method:", more_accurate_method)
```
